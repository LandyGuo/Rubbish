```python
import numpy as pd
import os
```

```python
#将正面的(编号为0)表情avi文件拷贝到"emotion_data"目录下
os.mkdir("emotion_data")
g = os.walk("./data")  
BaseDir = "emotion_data"
emotions =["angry","sad","dislike","laugh","neutral","smile","surprise"]
  
for path,d,filelist in g:  
    for filename in filelist:  
        file_path =  os.path.join(path, filename) 
        if file_path.endswith('.avi'):
            l = file_path.split('_')[1]
            o = filename.split('.')[0].split('_')[-1]
            if l in emotions and o=='0':
                name = file_path.split('/')[-1]
                save_path = os.path.join(BaseDir,name)
                open(save_path,'wb+').write(open(file_path,'rb').read())
                
                
```

```python
#拆分emotion_data下的视频帧为图片,保存在emotion images目录下
#调用系统命令进行拆帧(ffmpeg需事先安装):ffmpeg -i 001_angry_0.avi  -r 25 -f image2  001_angry_0_%3d.png
#每隔1s抽取3帧
CMD_TEMPLATE = "ffmpeg  -i {input_path} -r 3 -y  -f image2  {out_path}"
SAVE_DIR = "./emotion_images"
for path,dire,filelist in os.walk('./emotion_video'):
    for filename in filelist:
        if filename.endswith('.avi'):
            input_path = os.path.join(path, filename) 
            out_name = filename.split('.')[0]+"_%3d.png"
            out_path = os.path.join(SAVE_DIR,out_name)
            #拆分当前文件
            print "input_path:{} out_path:{}".format(input_path,out_path)
            os.system(CMD_TEMPLATE.format(**locals()))
            
```
